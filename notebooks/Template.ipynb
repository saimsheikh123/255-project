```python
# Looking at specific number of days after a speech to analyze movements
post_speech_window = 5

df["target_SP500"] = df["SP500"].pct_change(post_speech_window).shift(-1 * post_speech_window)   # 5-day window after speech
df["target_DJIA"] = df["DJIA"].pct_change(post_speech_window).shift(-1 * post_speech_window)   # 5-day window after speech
df["target_RSL"] = df["RSL"].pct_change(post_speech_window).shift(-1 * post_speech_window)   # 5-day window after speech
df["target_NDQ"] = df["NDQ"].pct_change(post_speech_window).shift(-1 * post_speech_window)   # 5-day window after speech
```

```python
feature_cols = [
    "Positive", "Negative", "Polarity",
    "topic_confidence","hawk_count", "dove_count",
    "HDI_dict", "fomc_hawk", "fomc_dove", "fomc_neutral"

] + [c for c in df.columns if c.startswith("topic_prob_")]

X = df[feature_cols]
```
