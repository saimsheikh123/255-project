```python
import yfinance as yf
import pandas as pd

# Define tickers for the indices
tickers = {
    "SP500": "^GSPC",
    "RSL": "^RUT",
    "DJIA": "^DJI",
    "NDQ": "^NDX"
}

start_date = "1996-01-01"
end_date = "2020-07-01"

# Download daily data for all indices
data = yf.download(
    list(tickers.values()),
    start=start_date,
    end=end_date,
    interval="1d",
    auto_adjust=False
)

# data has a multi-index (columns: [Open, High, Low, Close, Adj Close, Volume] per ticker)

adj_close = data["Adj Close"].copy()
adj_close.columns = tickers.keys()

print(adj_close.head())

adj_close.to_csv("index_prices_1996_2020.csv")
```
